```python
#(c) To create an element-wise comparison (greater, greater_equal,
#less and less_equal,equal, equal within a tolerance) of two given arrays.
#greater
a = np.array([10, 20, 30])
b = np.array([15, 20, 25])

print("Greater:", np.greater(a, b))            # [False False  True]
print("Greater Equal:", np.greater_equal(a, b))# [False  True  True]
print("Less:", np.less(a, b))                  # [ True False False]
print("Less Equal:", np.less_equal(a, b))      # [ True  True False]
print("Equal:", np.equal(a, b))                # [False  True False]

# Check if arrays are almost equal (within tolerance)
c = np.array([1.00001, 2.00001])
d = np.array([1.00002, 2.00002])
print("Allclose:", np.allclose(c, d))          # True (within default tolerance)
```

```python
#3.2 NumPy Program Using Methods
#ðŸ”¹ a. Extract numbers from array which are less and greater than a specified number
import numpy as np
arr = np.array([5, 15, 25, 35, 45])

# Extract numbers greater than 20
print("Greater than 20:", arr[arr > 20])  # [25 35 45]
print(arr)
# Extract numbers less than 30
print("Less than 30:", arr[arr < 30])     # [5 15 25]
print(arr)
```

```python
#Find indices of max and min numbers along a given axis
arr2d = np.array([[10, 20, 30], [5, 25, 15]])

# Index of max along axis 0 (column-wise)
print("Argmax (axis=0):", np.argmax(arr2d, axis=0))  # [0 1 0]

# Index of min along axis 1 (row-wise)
print("Argmin (axis=1):", np.argmin(arr2d, axis=1))  # [0 0]
```
