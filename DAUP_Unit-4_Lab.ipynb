```python
#4.1) Pandas DataSeries:
#i)Write a Pandas program to create and display a one-dimensional array-like object
#containing an array of data using Pandas module.

import pandas as pd

# Create a Pandas Series
data = pd.Series([1, "2", 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])

# Display the Pandas Series
print(data)
```

```python
#ii) Convert Pandas Series to Python List and its Type

import pandas as pd

# Create a Pandas Series
data = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])

# Convert Pandas Series to Python List
list_data = data.tolist()

# Print the list
print("List:", list_data)

# Print the type of list
print("Type of list:", type(list_data))

# Convert Pandas Series index and values to Python List
list_index = data.index.tolist()
list_values = data.values.tolist()

# Print the lists
print("List of Index:", list_index)
print("List of Values:", list_values)
```
